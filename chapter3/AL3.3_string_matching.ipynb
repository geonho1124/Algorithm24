# 1.문제 정의
문제이름 : 문자열 매칭   
문제설명 : 입력과 패턴의 문자가 일치하는지 문자를 비교하며 찾는다.   
문제예시 : 'T' = HELLO WORLD / 'P' = LO 라고하면 T에서 P(=LO)를 찾으면 매칭성공, 찾지못하면 실패이다.
# 2.알고리즘 설명
1. 문자열'T'와 패턴'P'를 입력받는다.
2. 변수 n,m에 각각 T와P의 길이를 저장한다.
3. for문을 통해 T를 기준으로 P와 일치하는지 문자를 비교한다.
4. T의 문자열에 P가 있으면 위치를 반환하고, 없을경우 -1을 반환한다.

# 4.알고리즘 개요
명칭 : 문자열 매칭   
입력변수 : 문자열 T, 패턴 P   
출력 : 패턴'P'을 기준으로 문자열'T'을 비교하여 둘 다 문자가 일치한 경우 패턴의 위치, 실패할 경우 -1 반환
# 5.알고리즘 코드

```python
def string_matching( T, P ):
    n = len(T)
    m = len(P)
    for i in range(n-m+1):
        j = 0
        while j < m and P[j]==T[i+j]:
            j = j + 1
            if j == m:
                return i
    return -1
```

# 6.테스트 코드

```python
def string_matching( T, P):
    n = len(T)
    m = len(P)
    for i in range(n-m+1):
        j=0
        while j < m and P[j]==T[i+j]:
            j = j + 1
            if j==m:
                return i
    return -1
        
text = 'HELLO WORLD'
pattern = 'LO'

print( string_matching(text, pattern))
```

# 8.복잡도 분석
처음으로 외부 루프 for i in range(n-m+1)에서 n-m+1번 반복하게 되고, 내부 루프에서는 모든문자를 비교하면 m번 반복한다.   
복잡도 T(n,m) = m(n-m+1)이므로 시간 복잡도는 O(nm)이 된다.
