```python
import math

def distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i],p[j])
            if dist < mindist:
                mindist = dist
    return mindist

p = [(2,3), (12,30), (40,50), (5,1), (12,10), (3,4)]
print(" :", closest_pair(p))
```
