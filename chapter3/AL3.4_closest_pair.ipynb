문제예시 : 'p'에 점의 좌표들을 입력받아 배열 내의 모든 점의 쌍을 비교하여

# 2.알고리즘 설명
1. 'p'에 각 점들의 좌표값을 입력받는다.
2. mindist에 최소 거리를 저장해야하므로 float("inf")를 사용해 무한대의 실수값을 저장한다.
3. 첫번째 for문을 통해 i부터 끝에서 두번째까지 반복한다.

5. dist에 유클리드 거리계산함수를 이용하여 거리를 계산한다.
6. for문을 돌며 dist가 mindist값보다 작을경우 dist값을 mindist에 저장한다.
7. 최종적으로 mindist에 최근접 쌍의 거리가 저장된다.

# 4.알고리즘 개요
명칭 : 최근접 쌍의 거리 구하기   
입력변수 : 점들의 좌표 값 p   
출력 : 모든 점들의 거리를 비교하여 가장 가까운 쌍의 거리 mindist
# 5.알고리즘 코드

```python
def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i],p[j])
            if dist < mindist:
                mindist = dist
return mindist
```

# 6.테스트 코드

```python
import math

def distance(point1, point2):
    x1, y1 = point1
    x2, y2 = point2
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i],p[j])
            if dist < mindist:
                mindist = dist
    return mindist

p = [(2,3), (12,30), (40,50), (5,1), (12,10), (3,4)]
print("최근접 거리:", closest_pair(p))
```
