# 1.문제 정의
문제이름 : 깊이 우선 탐색   
문제설명 : 시작 노드부터 더 이상 진행할 수 없을 때까지 탐색했다가 돌아와서 다른 방향을 탐색하여 모든 노드를 탐색한다.   
문제예시 : 'mygraph'에 입력받은 인접 리스트(딕셔너리와 집합)에 따라 시작 노드부터 최대치로 탐색을 시작하여 모든 노드를 탐색한다.
# 2.알고리즘 설명
1. 'mygraph'에 인접 리스트를 입력받는다.
2. dfs함수에 graph는 인접 리스트로 표현되고, start는 시작노드, visited는 이미 방문한 집합이다.
3. start가 visited에 포함되어 있지 않은지 검사한다.
4. visited에 start가 포함되어 있지 않다면 visited에 추가하여 노드를 방문했다고 표시한다.
5. print를 통해 start를 방문했다고 출력을 해준다.
6. nbr은 {인접정점} - {방문정점}이다.
7. for문을 통해 v에 대해 dfs(v)를 반복해서 호출한다.

# 4.알고리즘 개요
명칭 : 깊이 우선 탐색
입력변수 : 인접 리스트인 mygraph
출력 : 시작노드에서 최대치로 탐색했다가 돌아와 다른 방향으로 모든 노드를 탐색한 순서 dfs함수
# 5.알고리즘 코드

```python
def dfs( graph, start, visited ):
    if start not in visited :
        visited.add(start)
        print(start, end=' ')
        nbr = graph[start] - visited
        for v in nbr:
            dfs(graph, v, visited)
```

# 6.테스트 코드

```python
def dfs( graph, start, visited ):
    if start not in visited :
        visited.add(start)
        print(start, end=' ')
        nbr = graph[start] - visited
        for v in nbr:
            dfs(graph, v, visited)

mygraph = { "A" : {"B","C"},
           "B" : {"A","D"},
           "C" : {"A","D","E"},
           "D" : {"B","C","F"},
           "E" : {"C","G","H"},
           "F" : {"D"},
           "G" : {"E","H"},
           "H" : {"E","G"}
           }

print('DFS : ',end ='')
dfs(mygraph, "A",set() )
print()
```
