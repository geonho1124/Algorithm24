# 1.문제 정의
문제이름 : 분할 가능한 배낭 채우기(탐욕적 기법)   
문제설명 : 배낭 채우기 문제를 단위 무게당 가격이 높은 물건부터 넣는 탐욕적 기법을 사용하여 채운다.   
문제예시 : 물건들의 리스트 obj와 배낭의 용량 W를 입력받아 탐욕적 기법을 사용하여 배낭을 채운다.
# 2.알고리즘 설명
1. 물건들의 리스트 obj(튜플)와 배낭의 용량 W를 입력받는다.
2. obj를 내림차순으로 정렬하는데, 튜플에서 (가격 항) / (무게 항)의 값이 정렬 기준이다.
3. for루프를 돌며 용량이 다 찬 경우 루프를 빠져나온다.
4. 물건 전체가 들어갈 수 있는 경우 W에서 o[1]값(무게 항)을 빼주고 totalValue에 o[2]값(가격 항)을 더한다.
5. 물건의 일부만 넣을 수 있는 경우 W에서 O[1]을 나눠 fraction(배낭에 담을 수 있는 물건의 비율)에 저장한다.
6. o[2](가격 항)를 fraction(배낭에 담을 수 있는 물건의 비율)과 곱해서 totalValue에 더하여 저장한다.
7. o[1] * fraction을 하여 배낭의 실제 무게를 전체무게 W에서 빼주고, 정수형으로 변환한다.

# 4.알고리즘 개요
명칭 : 분할 가능한 배낭 채우기(탐욕적 기법)   
입력변수 : 물건들의 리스트 obj, 배낭의 용량 W   
출력 : 탐욕적 기법으로 배낭을 채운 값 totalValue
# 5.알고리즘 코드

```python
def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)
    
    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W -= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))
            
    return totalValue
```

# 6.테스트 코드

```python
def knapSack_fractional_greedy(obj, W):
    obj.sort(key = lambda o: o[2]/o[1], reverse=True)
    
    totalValue = 0
    for o in obj :
        if W <= 0 : break
        if W - o[1] >= 0:
            W -= o[1]
            totalValue += o[2]
        else:
            fraction = W / o[1]
            totalValue += o[2] * fraction
            W = int(W - (o[1] * fraction))
            
    return totalValue

obj = [ ("A", 10, 80), ("B", 12, 120), ("C", 8, 60) ]
print("W = 18", obj)
print("부분적인배낭(18): ", knapSack_fractional_greedy(obj,18), end='\n\n')
```
