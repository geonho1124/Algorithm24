# 1.문제 정의
문제이름 : 거스름돈 최소화 알고리즘(탐욕적 기법)   
문제설명 : 액면가가 가장 큰 동전부터 사용하여 거스름돈을 계산한다.   
문제예시 : 거스름돈이 580원일 때 액면가가 가장 큰 동전부터 차례대로 계산하여 각 동전의 개수를 반환한다.
# 2.알고리즘 설명
1. coins는 모든 동전의 액면가가 내림차순으로 정렬된 리스트를 입력받고, V는 거스름돈의 액수를 입력받는다.
2. 동전의 개수를 저장할 리스트 count를 만든다.
3. coins리스트의 개수만큼 for루프를 돈다.
4. 거스름돈 액수에 conis[i]를 나눠(몫의 나눗셈) 현재 액면가의 최대 사용 가능 개수를 cnt에 저장한다.
5. 계산한 동전 개수 cnt를 count에 추가한다.
6. 이미 계산된 (cnt * coins[i])금액을 V에서 빼서 다시 V에 저장해준다.
7. 계산한 금액을 뺀 V에서 다시 for루프를 돌며 계산한다.
8. count리스트를 리턴한다.

# 4.알고리즘 개요
명칭 : 거스름돈 최소화 알고리즘(탐욕적 기법)   
입력변수 : 모든 동전의 액면가가 내림차순으로 정렬된 리스트 coins, 거스름돈의 액수 V   
출력 : 각 액면가 동전의 개수 리스트 count
# 5.알고리즘 코드

```python
def min_coins_greedy( coins, V ) :
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count
```

# 6.테스트 코드

```python
def min_coins_greedy( coins, V ) :
    count = []
    for i in range(len(coins)) :
        cnt = V // coins[i]
        count.append(cnt)
        V -= cnt * coins[i]
    return count

coins = [500, 100, 50, 10, 5, 1]
V = 580
print("잔돈 액수 : ", V)
print("동전 종류 : ", coins)
print("동전 개수 : ",min_coins_greedy( coins, V ))
```
