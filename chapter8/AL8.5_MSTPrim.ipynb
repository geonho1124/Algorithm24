# 1.문제 정의
문제이름 : Prim의 최소비용 신장트리 알고리즘   
문제설명 : Prim의 MST알고리즘을 사용하여 모든정점을 연결하는 트리 중 가중치 합이 최소인 트리를 찾는다.   
문제예시 : 정점 리스트 vertex와 인접 행렬 adj를 입력받아 모든정점을 연결하는 트리 중 가중치 합이 최소인 트리를 찾는다.
# 2.알고리즘 설명
1. 그래프의 정점 리스트 vertex와 인접 행렬 adj를 입력받는다.
2. 정점의 수를 visze에 저장한다.
3. 초기값을 무한(INF)으로 설정하고 최소 거리 값을 저장할 dist에 저장한다.
4. dist[0] 즉, 시작 정점은 0으로 설정한다.
5. MST에 포함되어있는지 여부를 나타내므로 초기값을 False로 하여 selected에 저장한다.
6. 모든 정점을 반복하여 for루프를 돈다.
7. 현재 MST에 포함되지 않은 정점 중 dist값이 가장 작은 정점을 찾고 u에 저장한다.
8. 정점 u를 MST에 포함시킨다.
9. 내부루프에서 모든 정점을 반복하여 for루프를 돈다.
10. 정점 u와 v 사이에 간선이 있을 경우
11. 정점 v가 포함되지 않고, u와 v 사이의 가중치가 dist[v]보다 작은경우(v로 가는 더 짧은 경로 찾음)에 dist[v]를 adj[u][v]로 값을 갱신해준다.

# 4.알고리즘 개요
명칭 : Prim의 최소비용 신장트리 알고리즘   
입력변수 : 그래프의 정점 리스트 vertex, 인접 행렬 adj   
출력 : Prim의 최소비용 신장트리 알고리즘을 이용하여 구한 인접행렬 MSTPrim(vertex, weight)
# 5.알고리즘 코드

```python
def MSTPrim(vertex, adj) :
    vsize = len(vertex)
    dist = [INF] * vsize
    dist[0] = 0
    selected = [False] * vsize
    
    for i in range(vsize) :
        u = getMinVertex(dist, selected)
        selected[u] = True
        print(vertex[u], end=':')
        print(dist)
        
        for v in range(vsize):
            if(adj[u][v] != None):
                if selected[v]==False and adj[u][v]< dist[v] :
                    dist[v] = adj[u][v]
```

# 6.테스트 코드

```python
INF = float('inf')

def getMinVertex(dist, selected) :
    minv = -1
    mindist = INF
    for v in range(len(dist)) :
        if not selected[v] and dist[v]<mindist :
            mindist = dist[v]
            minv = v
    return minv

def MSTPrim(vertex, adj) :
    vsize = len(vertex)
    dist = [INF] * vsize
    dist[0] = 0
    selected = [False] * vsize
    
    for i in range(vsize) :
        u = getMinVertex(dist, selected)
        selected[u] = True
        print(vertex[u], end=':')
        print(dist)
        
        for v in range(vsize):
            if(adj[u][v] != None):
                if selected[v]==False and adj[u][v]< dist[v] :
                    dist[v] = adj[u][v]
                    
vertex = ['A', 'B', 'C', 'D', 'E', 'F', 'G' ]
weight = [ [ None,   29, None, None, None,   10, None ],
           [   29, None,   16, None, None, None,   15 ],
           [ None,   16, None,   12, None, None, None ],
           [ None, None,   12, None,   22, None,   18 ],
           [ None, None, None,   22, None,   27,   25 ],
           [   10, None, None, None,   27, None, None ],
           [ None,   15, None,   18,   25, None, None ] ]
MSTPrim(vertex, weight)
```
