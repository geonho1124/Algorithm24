# 1.문제 정의
문제이름 : 허프만 트리 생성 알고리즘   
문제설명 : 허프만 트리를 생성하여 허프만 코드를 구한다.   
문제예시 : 각 라벨의 가중치 ferq, 각 빈도에 대응하는 문자 label을 입력받아 허프만 트리를 생성하여 허프만 코드를 구한다.
# 2.알고리즘 설명
1. 라벨의 개수를 n에 저장한다.
2. 힙을 저장할 리스트 h를 생성한다.
3. 반복문을 사용하여 빈도와 라벨을 튜플로 묶어 힙에 추가한다.
4. 힙의 크기가 1이 될 때까지 두개의 최소 빈도 요소를 계속해서 합한다.
5. heapq.heappop(h)를 사용하여 가장 작은 두개의 요소를 힙에서 제거하고, 빈도를 합친 새로운 노드를 힙에 추가한다.
6. 힙에 남은 마지막 요소가 허프만 트리를 나타낸다.

# 4.알고리즘 개요
명칭 : 허프만 트리 생성 알고리즘   
입력변수 : 각 라벨의 가중치 ferq, 각 빈도에 대응하는 문자 label   
출력 : 힙에 남은 마지막 요소 heapq.heappop(h)
# 5.알고리즘 코드

```python
import heapq
def make_heap_tree(freq, label):
    n = len(freq)
    h=[]
    for i in range(n):
        heapq.heappush(h, (freq[i], label[i]))
        
    for i in range(1, n) :
        e1 = heapq.heappop(h)
        e2 = heapq.heappop(h)
        heapq.heappush(h, (e1[0]+e2[0], e1[1]+e2[1]))
        print(e1, "+", e2)
        
    print(heapq.heappop(h))
```

# 6.테스트 코드

```python
import heapq
def make_heap_tree(freq, label):
    n = len(freq)
    h=[]
    for i in range(n):
        heapq.heappush(h, (freq[i], label[i]))
        
    for i in range(1, n) :
        e1 = heapq.heappop(h)
        e2 = heapq.heappop(h)
        heapq.heappush(h, (e1[0]+e2[0], e1[1]+e2[1]))
        print(e1, "+", e2)
        
    print(heapq.heappop(h))
    
label = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H' ]
freq = [24, 3, 8, 10, 33, 6, 4, 12]
make_heap_tree(freq, label)
```
