# 1.문제 정의
문제이름 : 병합 알고리즘(정렬된 두 리스트의 병합)   
문제설명 : 정렬된 왼쪽 리스트와 오른쪽 리스트를 하나의 리스트로 병합시킨다.   
문제예시 : 왼쪽리스트[1, 3, 7, 8]과 오른쪽 리스트[2, 4, 5, 9]를 병합시켜   
[1, 2, 3, 4, 5, 6, 7, 8, 9]로 병합시킨다.
# 2.알고리즘 설명
1. i가 mid보다 작거나 같고, j가 right보다 작거나 같을 때 반복된다.
2. A[i]가 A[j]보다 작거나 같을 때, A[i]값을 sorted[k]에 저장하고 i와 k를 1씩 증가시킨다.
3. A[i]가 A[j]보다 클 때, A[j]값을 sorted[k]에 저장하고 j와 k를 1씩 증가시킨다.
4. 그 후에 i가 mid보다 클 경우 남은 부분 리스트를 슬라이싱을 이용하여 복사해준다.
5. sorted를 원래 리스트(A)로 다시 복사한다.

# 4.알고리즘 개요
명칭 : 병합 알고리즘(정렬된 두 리스트의 병합)   
입력변수 : 정렬할 리스트 A, 정렬된 왼쪽 리스트 첫번째 인덱스 left,    리스트의 중간 mid, 정렬된 오른쪽 리스트 마지막 인덱스 right   
출력 : 정렬된 리스트 A
# 5.알고리즘 코드

```python
def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right :
        if A[i] <= A[j] :
            sorted[k] = A[i]
            i, k = i+1, k+1
        else:
            sorted[k] = A[j]
            j, k = j+1, k+1
            
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]
        
    A[left:right+1] = sorted[left:right+1]
```

#  6.테스트 코드

```python
def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right :
        if A[i] <= A[j] :
            sorted[k] = A[i]
            i, k = i+1, k+1
        else :
            sorted[k] = A[j]
            j, k = j+1, k+1
            
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]
        
    A[left:right+1] = sorted[left:right+1]

# AL(5.1) merge_sort함수
def merge_sort(A, left, right) :
    if left<right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)
        
data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
sorted = [0] * len(data)
merge_sort(data,0,len(data)-1)
print(data)
```
