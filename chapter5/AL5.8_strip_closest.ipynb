# 4.알고리즘 개요
명칭 : 띠 영역 내에서 d보다 작은 최근접 쌍의 거리 찾기   
입력변수 : 점들의 리스트 P, 최소거리 d   
출력 : 최근접 쌍의 거리 d_min
# 5.알고리즘 코드

```python
def strip_closest(P, d):
    n = len(P)
    d_min = d
    P.sort(key = lambda point: point[1])
    
    for i in range(n) :
        j = i + 1
        while j < n and (P[j][1] - P[i][1]) < d_min:
            dij = distance(P[i], P[j])
            if dij < d_min :
                d_min = dij
            j += 1
    return d_min
```

# 6.테스트 코드

```python
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def strip_closest(P, d):
    n = len(P)
    d_min = d
    P.sort(key = lambda point: point[1])
    
    for i in range(n) :
        j = i + 1
        while j < n and (P[j][1] - P[i][1]) < d_min:
            dij = distance(P[i], P[j])
            if dij < d_min :
                d_min = dij
            j += 1
    return d_min
# AL3.4
def closest_pair(p):
    n = len(p)
    mindist = float("inf")
    for i in range(n-1):
        for j in range(i+1, n):
            dist = distance(p[i], p[j])
            if dist < mindist:
                mindist = dist
    return mindist
# AL5.9
def closest_pair_dist(P, n) :
    if n <= 3:
        return closest_pair(P)
    
    mid = n // 2
    mid_x = P[mid][0]
    
    dl = closest_pair_dist(P[:mid], mid)
    dr = closest_pair_dist(P[mid:], n-mid)
    d = min(dl, dr)
    
    Pm = []
    for i in range(n):
        if abs(P[i][0] - mid_x) < d:
            Pm.append(P[i])
            
    ds = strip_closest(Pm, d)
    return ds

p = [(2,3), (12,30), (40,50), (5,1), (12, 10), (3, 4)]
p.sort(key = lambda point: point[0])
print("가장 가까운 두 점의 거리", closest_pair_dist(p, len(p)))
```
