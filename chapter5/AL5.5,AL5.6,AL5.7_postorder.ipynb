```python
# AL5.5_preorder
def preorder(n) :
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)
# AL5.6_inorder  
def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
# AL5.7_postorder
def postorder(n) :
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
```

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

d = TNode('D')
e = TNode('E')
b = TNode('B', d, e)
f = TNode('F')
c = TNode('C', f)
root = TNode('A', b, c)

# AL5.5_preorder
def preorder(n) :
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)
        # AL5.6_inorder  
def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
# AL5.7_postorder
def postorder(n) :
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')

print("inorder : ",end='')
inorder(root)
print()
print("preorder : ",end='')
preorder(root)
print()
print("postorder : ",end='')
postorder(root)
print()
```
