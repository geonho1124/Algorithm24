# 1.문제 정의
문제이름 : 이진트리의 전위/중위/후위 순회   
문제설명 : 트리의 노드들을 전위, 중위, 후위순회 방식으로 순회한다.   
문제예시 : 어떤 이진트리의 노드들을 전위, 중위, 후위순회 방식으로 순회하여 순서대로 표시한다.
# 2.알고리즘 설명
(전위 순위) - VLR방식으로 루트노드를 방문하고 왼쪽 서브 트리를 방문하고 오른쪽 서브 트리를 방문한다.   
(중위 순위) - LVR방식으로 왼쪽 서브 트리를 방문하고 루트노드를 방문하고 오른쪽 서브 트리를 방문한다.   
(후위 순위) - LRV방식으로 왼쪽 서브 트리를 방문하고 오른쪽 서브 트리를 방문하고 루트노드를 방문한다.

# 4.알고리즘 개요
명칭 : 이진트리의 전위/중위/후위 순회   
입력변수 : 이진트리와 root노드   
출력 : 각각의 노드를 전위, 중위, 후위 순회 방식으로 순회한 순서
# 5.알고리즘 코드

```python
# AL5.5_preorder
def preorder(n) :
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)
# AL5.6_inorder  
def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
# AL5.7_postorder
def postorder(n) :
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')
```

# 6.테스트 코드

```python
class TNode:
    def __init__(self, data, left=None, right=None):
        self.data = data
        self.left = left
        self.right = right

d = TNode('D')
e = TNode('E')
b = TNode('B', d, e)
f = TNode('F')
c = TNode('C', f)
root = TNode('A', b, c)

# AL5.5_preorder
def preorder(n) :
    if n is not None :
        print(n.data, end=' ')
        preorder(n.left)
        preorder(n.right)
        # AL5.6_inorder  
def inorder(n) :
    if n is not None :
        inorder(n.left)
        print(n.data, end=' ')
        inorder(n.right)
# AL5.7_postorder
def postorder(n) :
    if n is not None :
        postorder(n.left)
        postorder(n.right)
        print(n.data, end=' ')

print("inorder : ",end='')
inorder(root)
print()
print("preorder : ",end='')
preorder(root)
print()
print("postorder : ",end='')
postorder(root)
print()
```
