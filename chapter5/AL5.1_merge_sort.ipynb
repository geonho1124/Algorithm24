# 1.문제 정의
문제이름 : 병합 정렬

문제예시 : list=[8,1,7,3,9,2,4,5]를 입력받아 병합 정렬을 사용하여 리스트를 오름차순으로 정렬한다.
# 2.알고리즘 설명
1. 정렬할 리스트 A, 정렬이 시작되는 인덱스 left, 정렬이 끝나는 인덱스 right를 입력받는다.
2. 종료조건을 left가 right보다 크거나 같으면 종료되는 것으로 정한다.
3. left와 right를 더해 2를 나누고(정수의나눗셈) mid값에 저장한다.
4. left부터 mid까지(왼쪽 리스트)부분에서 merge_sort함수를 사용(리스트 나눔)한다.
5. mid+1부터 right까지(오른쪽 리스트)부분에서 merge_sort함수를 사용(리스트 나눔)한다.
6. 함수 merge(AL5.2)를 사용하여 정렬된 리스트를 병합하여 정렬된 리스트를 출력한다.

# 4.알고리즘 개요
명칭 : 병합 정렬   
입력변수 : 정렬할 리스트 A, 정렬이 시작되는 인덱스 left, 정렬이 끝나는 인덱스 right   
출력 : 정렬된 리스트 A
# 5.알고리즘 코드

```python
def merge_sort(A, left, right) :
    if left<right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)
```

# 6.테스트 코드

```python
# AL(5.2) merge함수
def merge(A, left, mid, right) :
    k = left
    i = left
    j = mid + 1
    while i<=mid and j<=right :
        if A[i] <= A[j] :
            sorted[k] = A[i]
            i, k = i+1, k+1
        else :
            sorted[k] = A[j]
            j, k = j+1, k+1
            
    if i > mid :
        sorted[k:k+right-j+1] = A[j:right+1]
    else :
        sorted[k:k+mid-i+1] = A[i:mid+1]
        
    A[left:right+1] = sorted[left:right+1]

def merge_sort(A, left, right) :
    if left<right :
        mid = (left + right) // 2
        merge_sort(A, left, mid)
        merge_sort(A, mid + 1, right)
        merge(A, left, mid, right)
        
data = [5, 3, 8, 4, 9, 1, 6, 2, 7]
sorted = [0] * len(data)
merge_sort(data,0,len(data)-1)
print(data)
```
