# 1.문제 정의
문제이름 : 0-1 배낭 채우기 알고리즘(분할 정복 기법)   
문제설명 : 분할 정복 기법을 사용하여 배낭 채우기 문제를 해결한다.   
문제예시 : 분할 정복 기법을 사용하여 주어진 값 val=[60, 100, 190, 120, 200, 150], wt=[2, 5, 8, 4, 7, 6], W=18, n=len(val)   
으로 배낭 채우기 문제를 해결한다.
# 2.알고리즘 설명
1. 기반상황으로 물건이 없거나 배낭 용량이 0일경우 0을 리턴한다.
2. n번째 항목이 배낭의 용량보다 클 경우 배낭에 넣을 수 없으므로 n-1로 다시 계산한다.
3. 그렇지 않을 경우, 항목을 넣지 않는 경우의 배낭 가치와 넣는 경우의 배낭 가치를 계산하고 각각 저장한다.
4. 둘 중 더 큰 값을 리턴한다.

# 4.알고리즘 개요
명칭 : 0-1 배낭 채우기 알고리즘(분할 정복 기법)   
입력변수 : 현재 배낭의 용량 W, 물건들의 무게를 저장한 리스트 wt, 물건들의 가치를 저장한 리스트 val, 물건들의 수 n   
출력 : 항목을 넣지 않는 경우의 배낭 가치와 넣는 경우의 배낭 가치를 계산한 값들 중 더 큰 값 max(valWith, valWithout)
# 5.알고리즘 코드

```python
def knapSack_dc(W, wt, val, n):
    if n == 0 or W == 0 :
        return 0
    
    if (wt[n-1] > W):
        return knapSack_dc(W, wt, val, n-1)
    else:
        valWithout = knapSack_dc(W, wt, val, n-1)
        valWith = val[n-1] + knapSack_dc(W-wt[n-1], wt, val, n-1)
        return max(valWith, valWithout)
```

# 6.테스트 코드

```python
def knapSack_dc(W, wt, val, n):
    if n == 0 or W == 0 :
        return 0
    
    if (wt[n-1] > W):
        return knapSack_dc(W, wt, val, n-1)
    else:
        valWithout = knapSack_dc(W, wt, val, n-1)
        valWith = val[n-1] + knapSack_dc(W-wt[n-1], wt, val, n-1)
        return max(valWith, valWithout)
    
val=[60, 100, 190, 120, 200, 150]
wt=[2, 5, 8, 4, 7, 6]
W=18
n=len(val)

print("결과 = ",knapSack_dc(W, wt, val, n))
```
