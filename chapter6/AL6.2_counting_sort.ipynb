# 1.문제 정의
문제이름 : 카운팅 정렬   
문제설명 : 리스트의 각 항목들의 수를 세는 방법으로 정렬한다.   
문제예시 : [1, 4, 1, 2, 7, 5, 2]를 카운팅 정렬을 사용하여 정렬시킨다.
# 2.알고리즘 설명
1. 정렬 결과를 저장하기 위해 0으로 초기화된 output리스트를 만들어준다.
2. 숫자의 빈도를 저장하기 위해 0으로 초기화된 count리스트를 만들어준다.
3. 첫번째 for루프를 돌며 숫자별로 빈도를 계산한다.
4. 두번째 for루프를 돌며 count[i]가 해당하는 숫자의 위치가 되도록 수정한다.
5. 세번째 for루프를 돌며 모든 입력 A[i]를 해당위치 count[A[i]]-1에 저장하고 킷값의 위치를 하나 줄인다.
6. 정렬 결과를 원래의 리스트에 복사한다.

# 4.알고리즘 개요
명칭 : 카운팅 정렬   
입력변수 : 정렬할 리스트 A   
출력 : 카운팅 정렬된 리스트 A
# 5.알고리즘 코드

```python
def counting_sort(A):
    output = [0] * len(A)
    count = [0] * MAX_VAL
    
    for i in A:
        count[i] += 1
        
    for i in range(MAX_VAL):
        count[i] += count[i-1]
        
    for i in range(len(A)):
        output[count[A[i]]-1] = A[i]
        count[A[i]] -= 1
        
    for i in range(len(A)):
        A[i] = output[i]
```

# 6.테스트 코드

```python
def counting_sort(A):
    output = [0] * len(A)
    count = [0] * MAX_VAL
    
    for i in A:
        count[i] += 1
        
    for i in range(MAX_VAL):
        count[i] += count[i-1]
        
    for i in range(len(A)):
        output[count[A[i]]-1] = A[i]
        count[A[i]] -= 1
        
    for i in range(len(A)):
        A[i] = output[i]
        
MAX_VAL = 10
data = [1, 4, 1, 2, 7, 5, 2]
counting_sort(data)
print("정렬된 리스트 : ", data)
```
