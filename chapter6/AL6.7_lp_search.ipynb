```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0 and (table[id] != None):
        id = (id + 1 + M) % M
        count -= 1
    if count > 0:
        table[id] = key
    return

keys = [45, 27, 88, 9, 71, 60, 46, 38, 24]

for key in keys:
    lp_insert(key)

def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :
            return None
        if table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
    return None

print("46 : ", lp_search(46))
print("39 : ", lp_search(39))
```
