# 1.문제 정의
문제이름 : 선형 조사법의 탐색 알고리즘   
문제설명 : 탐색할 탐색 키를 입력하여 키값이 테이블에 있는지 탐색한다.   
문제예시 : 선형 조사법으로 테이블에 46과 39가 있는지 탐색한다.
# 2.알고리즘 설명
1. 키값을 해시함수를 사용하여 id에 저장한다.
2. 테이블의 크기인 M을 count에 저장한다.
3. count가 0보다 클 경우 while루프를 돈다.
4. table의 주소(버킷)가 비어있으면 None을 리턴한다.(key가 테이블에 없음)
5. table의 주소(버킷)에 키값이 있으면 값을 리턴한다.
6. 값이 없는 경우에는 다음 칸을 탐색한다.
7. count값을 감소시키고 다시 while루프를 돈다.
8. 루프가 종료될 때까지 성공하지 못하면 None을 리턴한다.(key가 해시 테이블에 없음)

# 4.알고리즘 개요
명칭 : 선형 조사법의 탐색 알고리즘   
입력변수 : 탐색할 탐색 키값 key   
출력 : 탐색이 성공할 경우 key값, 실패할 경우 None
# 5.알고리즘 코드

```python
def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :
            return None
        if table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
    return None
```

# 6.테스트 코드

```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0 and (table[id] != None):
        id = (id + 1 + M) % M
        count -= 1
    if count > 0:
        table[id] = key
    return

keys = [45, 27, 88, 9, 71, 60, 46, 38, 24]

for key in keys:
    lp_insert(key)

def lp_search(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None :
            return None
        if table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
    return None

print("46탐색 : ", lp_search(46))
print("39탐색 : ", lp_search(39))
```
