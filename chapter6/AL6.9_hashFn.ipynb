# 1.문제 정의
문제이름 : 문자열 탐색키의 해시 함수 계산   
문제설명 : 해시주소를 저장할 키(문자열)를 입력받아 해시주소를 리턴한다.   
문제예시 : 키(문자열)를 입력받아 키값의 해시주소를 출력한다.
# 2.알고리즘 설명
1. 문자열의 각 문자에 대한 아스키 코드 값의 합을 저장하는 sum에 초기값을 0으로 설정한다.
2. 입력된 문자열 key의 각 문자를 순회하며 문자 c값을 아스키 코드 값으로 변환시켜 sum과 더하여 sum에 다시 저장한다.
3. sum을 테이블 크기로 나눈 나머지를 리턴한다.

# 4.알고리즘 개요
명칭 : 문자열 탐색키의 해시 함수 계산   
입력변수 : 문자열 key   
출력 : 문자열 key의 해시주소
# 5.알고리즘 코드

```python
def hashFn(key) :
    sum = 0
    for c in key :
        sum = sum + ord(c)
    return sum % M
```

# 6.테스트 코드

```python
def hashFn(key) :
    sum = 0
    for c in key :
        sum = sum + ord(c)
    return sum % M

M = 13

keys = ["apple", "mango", "banana"]

for key in keys :
    print(f"키: {key}, 해시주소: {hashFn(key)}")
```
