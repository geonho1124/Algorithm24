```python
#(AL6.3)
NO_OF_CHARS = 128
def shift_table(pat):
    m = len(pat)
    tbl = [m]*NO_OF_CHARS
    
    for i in range(m-1):
        tbl[ord(pat[i])] = m-1-i
        
    return tbl
```

```python
#(AL6.4)
def search_horspool(T, P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1
    while(i <= n-1):
        k = 0
        while k <= m-1 and P[m-1-k]==T[i-k]:
            k += 1
        if k == m :
            return i-m+1
        else :
            tc = t[ord(T[i-k])]
            i += (tc-k)
    return -1
```
