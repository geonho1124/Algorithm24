# 1.문제 정의
문제이름 : 호스풀 알고리즘(시프트 테이블)   
문제설명 : 텍스트에서 패턴을 뒤에서부터 탐색하여 문자가 있는지 검사한다.   
문제예시 : 텍스트 "APPLEMANGOBANANAGRAPE"에서 패턴 "BANANA"이 위치를 찾아 인덱스를 반환한다.
# 2.알고리즘 설명
### (AL6.3)
1. 패턴의 길이를 m에 저장해준다.
2. 시프트 테이블을 만들기위해 NO_OF_CHARS길이만큼 m으로 채운 리스트를 tbl에 저장한다.(모든 문자는 패턴에 없다고 생각)
3. 패턴의 모든 문자에 대해서 알파벳이 몇번째 문자인지 ord함수를 통해 변환하여 tbl에 저장하고,   
패턴에 중복된 문자가 있을 경우 맨 오른쪽 문자에 의한 거리가 들어간다.
### (AL6.4)
1. 패턴의 길이를 m에 저장한다.
2. 텍스트의 길이를 n에 저장한다.
3. 시프트 테이블을 생성하여 t에 저장한다.
4. 패턴의 가장 오른쪽에 위치한 위치를 i에 저장한다.
5. while문으로 모든 위치에 대해서 검사를 진행한다.
6. 매칭된 개수를 저장하기위해 k값에 0을 저장한다.
7. while문을 사용하여 패턴을 뒤에서 앞 방향으로 맞지 않을 때까지 검사한다.
8. 패턴이 맞을 경우 k를 1씩 증가시켜 계속 앞으로 검사를 진행시킨다.
9. 만약 패턴의 모든 문자가 일치한 경우 해당 위치를 반환시킨다.
10. 패턴의 모든 문자가 일치하지 않을 경우 시프트 테이블을 이용해 건너뛴다.
11. 매칭을 실패하면 -1을 리턴한다.

# 4.알고리즘 개요
명칭 : 호스풀 알고리즘(시프트 테이블)   
입력변수 : 텍스트 T, 패턴 P   
출력 : 텍스트에서 패턴의 위치
# 5.알고리즘 코드

```python
#(AL6.3)
NO_OF_CHARS = 128
def shift_table(pat):
    m = len(pat)
    tbl = [m]*NO_OF_CHARS
    
    for i in range(m-1):
        tbl[ord(pat[i])] = m-1-i
        
    return tbl
```

```python
#(AL6.4)
def search_horspool(T, P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m-1
    while(i <= n-1):
        k = 0
        while k <= m-1 and P[m-1-k]==T[i-k]:
            k += 1
        if k == m :
            return i-m+1
        else :
            tc = t[ord(T[i-k])]
            i += (tc-k)
    return -1
```

# 6.테스트 코드
