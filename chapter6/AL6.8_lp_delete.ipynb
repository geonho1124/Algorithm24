# 1.문제 정의
문제이름 : 선형 조사법의 삭제 알고리즘   
문제설명 : 테이블에서 삭제하려는 키 값을 입력받아 삭제시킨다.   
문제예시 : 테이블에서 60을 삭제시킨다.
# 2.알고리즘 설명
1. 삭제하려는 키값에 해시함수를 사용하여 id에 저장한다.
2. 테이블의 크기 M을 count에 저장한다.
3. count가 0보다 클 경우 while문을 반복한다.
4. 현재 테이블이 None일 경우 해당 키값이 테이블에 존재하지 않기 때문에 None을 리턴한다.
5. 현재 테이블이 -1(삭제된 항목)이 아니고 key값과 일치하면 현재 테이블에 -1을 저장시켜 키값을 삭제시킨다.
6. 선형 탐사를 하기위해 id(주소값)을 1증가시킨다.
7. count값(시도 횟수)을 1감소시킨다.

# 4.알고리즘 개요
명칭 : 선형 조사법의 삭제 알고리즘   
입력변수 : 삭제시키려는 키값 key   
출력 : 테이블에 키값이 있는 경우 -1저장, 없는경우 None
# 5.알고리즘 코드

```python
def lp_delete(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None : return
        if table[id] != -1 and table[id] == key :
           table[id] = -1
           return
        id = (id + 1 + M) % M
        count -= 1
```

# 6.테스트 코드

```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0 and (table[id] != None):
        id = (id + 1 + M) % M
        count -= 1
    if count > 0:
        table[id] = key
    return

keys = [45, 27, 88, 9, 71, 60, 46, 38, 24]

for key in keys:
    lp_insert(key)

def lp_delete(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None : return
        if table[id] != -1 and table[id] == key :
           table[id] = -1
           return
        id = (id + 1 + M) % M
        count -= 1
        
lp_delete(60)
print("테이블에 60을 삭제한 결과 : ", table)
```
