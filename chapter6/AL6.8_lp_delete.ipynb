```python
M = 13
table = [None] * M

def hashFn(key):
    return key % M

def lp_insert(key):
    id = hashFn(key)
    count = M
    while count > 0 and (table[id] != None):
        id = (id + 1 + M) % M
        count -= 1
    if count > 0:
        table[id] = key
    return

keys = [45, 27, 88, 9, 71, 60, 46, 38, 24]

for key in keys:
    lp_insert(key)

def lp_delete(key) :
    id = hashFn(key)
    count = M
    while count>0 :
        if table[id] == None : return
        if table[id] != -1 and table[id] == key :
           table[id] = -1
           return
        id = (id + 1 + M) % M
        count -= 1
        
lp_delete(60)
print(" 60   : ", table)
```
