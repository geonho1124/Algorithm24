# 1.문제 정의
문제이름 : 거듭제곱(억지 기법)   
문제설명 : 억지 기법을 사용하여 거듭제곱을 구한다.   
문제예시 : x=5, n=3일때, for루프를 n번 반복하여 x를 곱한다.
# 2.알고리즘 설명
1. 곱할 값 x와 반복할(제곱) n을 입력받는다.
2. for루프를 통하여 n-1번 반복한다.
3. result = result * x를 n번 반복하여 거듭제곱을 구하여 결과값으로 result를 반환한다.
# 3.손으로 푼 예제

# 4.알고리즘 개요
명칭 : 거듭제곱(억지 기법)   
입력변수 : 곱할 값 x, 반복할(제곱) 값 n   
출력 : x를 n번 곱한(제곱한) result
# 5.알고리즘 코드

```python
def slow_power(x, n) :
    result = 1.0
    for i in range(n) :
        result = result * x
    return result
```

# 6.테스트 코드

```python
def slow_power(x, n) :
    result = 1.0
    for i in range(n) :
        result = result * x
    return result

print(slow_power(5,3))
```
