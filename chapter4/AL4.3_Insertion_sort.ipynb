# 2.알고리즘 설명
1. 리스트 A를 입력받는다.
2. 리스트 A의 길이를 n에 저장한다.
3. for문을 통해 1(두번째)부터 n-1까지 반복한다.(첫번째는 이미 정렬)
4. 리스트 A의[인덱스 i]의 값을 key에 저장한다.
5. key값의 전 인덱스부터 정렬된 리스트를 탐색하기 위해 j를 i - 1로 초기화한다.
6. while문을 통해 key값보다 큰 값을 찾았을 때, j+1을하여 오른쪽으로 한칸씩 이동시킨다.
7. j를 j - 1하여 정렬된 리스트를 탐색한다.
8. key값 보다 작은 값이거나 정렬된 리스트의 시작에 도달하면 key값을 그 위치에 삽입한다.

```python
def insertion_sort(A) :
    n = len(A)
    for i in range(1, n) :
        key = A[i]
        j = i - 1
        while j>=0 and A[j] > key :
            A[j + 1] = A[j]
            j = j - 1
            A[j + 1] = key
            printstep(A, i)
```

```python
def insertion_sort(A) :
    n = len(A)
    for i in range(1, n) :
        key = A[i]
        j = i - 1
        while j>=0 and A[j] > key :
            A[j + 1] = A[j]
            j = j - 1
            A[j + 1] = key
    return A

list_A = [ 5, 3, 8, 4, 9, 1, 6, 2, 7 ]
print(insertion_sort(list_A))
```
