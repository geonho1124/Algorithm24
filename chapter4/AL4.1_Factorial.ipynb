# 1.문제 정의
문제이름 : 팩토리얼(하향식 축소 정복 기법)   
문제설명 : 하향식 축소 정복 기법을 사용하여 팩토리얼을 구한다.   
문제예시 : 만약 n=3이라고 하면, n이 1이 될 때까지 재귀적으로   
(n * factorial_recur(n-1))을 반복하여 하향식 축소 정복 기법으로 3 * 2 * 1 = 6을 구하게된다.
# 2.알고리즘 설명
1. n의 값에 따라 1일 경우 1을 리턴한다.
2. n의 값이 1이 아닐경우 (n * factorial_recur(n-1))이 1까지 반복된다.
3. 따라서 n-1에 대해 재귀적으로 함수를 호출하여 그 결과를 다시 반복한다.
4. 최종적으로 n에서 1이 될 때까지 반복하다 1이 되면 1을 리턴하고 종료된다.
5. n부터 1까지의 수를 곱하여 결과를 반환한다.

# 4.알고리즘 개요
명칭 : 팩토리얼(하향식 축소 정복 기법)   
입력변수 : 팩토리얼의 값을 정하는 n   
출력 : 팩토리얼을 하향식 축소 정복 기법을 통해 구한 factorial_recur함수
# 5.알고리즘 코드

```python
def factorial_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n * factorial_recur(n - 1))
```

# 6.테스트 코드

```python
def factorial_recur(n) :
    if n == 1 :
        return 1
    else :
        return (n * factorial_recur(n - 1))
    
print('n=3일때',factorial_recur(3))
```
