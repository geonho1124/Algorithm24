# 1.문제 정의
문제이름 : 이진 탐색(순환구조)   
문제설명 : 정렬된 리스트A에서 key값을 찾아 탐색을 성공하면 인덱스를 반환하고, 실패하면 -1을 반환한다.   
문제예시 : 정렬된 리스트 [1, 3, 8, 13, 13, 16, 21, 26, 27, 30]을 입력받아 key값 16을 찾아 인덱스 5를 반환한다.
# 2.알고리즘 설명
1. 정렬된 리스트 A, 찾을 값 key, 탐색범위시작 인덱스 low, 탐색범위끝 인덱스 high를 입력받는다.
2. 탐색범위 시작인덱스 low와 탐색범위끝인덱스 high를 더해 2를 나눠(정수나눗셈) 중간값 mid를 구한다.
3. key값과 A[mid]가 일치할 경우 인덱스 mid를 반환한다.
4. key값보다 A[mid]가 클 경우, 배열의 왼쪽(즉, A[mid]보다 작은 값)에서 이진 탐색을 재귀적으로 수행한다.
5. key값보다 A[mid]가 작을 경우, 배열의 오른쪽(즉, A[mid]보다 큰 값)에서 이진 탐색을 재귀적으로 수행한다.
6. 함수를 재귀적으로 수행하여 key값과 A[mid]가 일치할 때 mid값(인덱스값)을 리턴한다.
7. 만약 low가 high보다 크면 key값을 찾을 수 없기때문에 -1을 반환한다.

# 4.알고리즘 개요
명칭 : 이진 탐색(순환구조)   
입력변수 : 정렬된 리스트 A, 찾을 값 key, 탐색범위시작 인덱스 low, 탐색범위끝 인덱스 high   
출력 : 찾을 값 key의 인덱스인 mid
# 5.알고리즘 코드

```python
def binary_search(A, key, low, high) :
    if (low <= high) :
        mid = (low + high) // 2
        if key == A[mid] :
            return mid
        elif key < A[mid] :
            return binary_search(A, key, low, mid-1)
        else :
            return binary_search(A, key, mid+1, high)
    return -1
```

# 6.테스트 코드

```python
def binary_search(A, key, low, high) :
    if (low <= high) :
        mid = (low + high) // 2
        if key == A[mid] :
            return mid
        elif key < A[mid] :
            return binary_search(A, key, low, mid-1)
        else :
            return binary_search(A, key, mid+1, high)
    return -1

listA = [1, 3, 8, 13, 13, 16, 21, 26, 27, 30]
print("key = 16","인덱스 값은=",binary_search(listA, 16, 0, len(listA)-1))
```
